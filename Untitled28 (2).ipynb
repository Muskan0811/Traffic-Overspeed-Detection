```python
import os
import cv2

video_dir = "/content/dataset2"
output_root = "/content/output"

os.makedirs(output_root, exist_ok=True)

frame_skip = 2 # take every 5th frame

for video_name in os.listdir(video_dir):
    if not video_name.endswith(".mp4"):
        continue

    video_path = os.path.join(video_dir, video_name)
    cap = cv2.VideoCapture(video_path)

    count = 0
    saved = 0

    video_id = video_name.split(".")[0]

    # 👉 Create a separate folder for each video
    video_output_dir = os.path.join(output_root, video_id)
    os.makedirs(video_output_dir, exist_ok=True)

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if count % frame_skip != 0:
            count += 1
            continue

        frame = cv2.resize(frame, (256, 256))

        filename = f"{video_id}_frame_{saved:05d}.jpg"
        cv2.imwrite(os.path.join(video_output_dir, filename), frame)

        saved += 1
        count += 1

    cap.release()
    print(f"{video_name} → {saved} frames extracted into {video_output_dir}")

print("All videos processed")
```

```python
import os
import shutil

input_root = "/content/output"
output_folder = "/content/all_frames"

os.makedirs(output_folder, exist_ok=True)

count = 0

for folder in os.listdir(input_root):
    folder_path = os.path.join(input_root, folder)

    if not os.path.isdir(folder_path):
        continue

    for img in sorted(os.listdir(folder_path)):

        src = os.path.join(folder_path, img)

        new_name = f"frame_{count:06d}.jpg"
        dst = os.path.join(output_folder, new_name)

        shutil.copy(src, dst)

        count += 1

print("Done. Total frames:", count)
print("Saved in:", output_folder)
```

```python
import os
import cv2

input_folder = "/content/all_frames"
output_folder = "/content/denoised_frames"

os.makedirs(output_folder, exist_ok=True)

count = 0

for img_name in sorted(os.listdir(input_folder)):

    img_path = os.path.join(input_folder, img_name)
    img = cv2.imread(img_path)

    if img is None:
        continue

    # 🧼 DENOISING
    denoised = cv2.fastNlMeansDenoisingColored(
        img,
        None,
        h=10,
        hColor=10,
        templateWindowSize=7,
        searchWindowSize=21
    )

    save_path = os.path.join(output_folder, img_name)
    cv2.imwrite(save_path, denoised)

    count += 1

print("✅ Denoising completed")
print("Total frames processed:", count)
print("Saved in:", output_folder)
```

```python
import os
import cv2
from ultralytics import YOLO

# load better model
model = YOLO("yolov8m.pt")

input_folder = "/content/denoised_frames"
output_folder = "/content/detected_frames"
os.makedirs(output_folder, exist_ok=True)

# vehicle class IDs (COCO)
vehicle_classes = [2, 3, 5, 7]  # car, bike, bus, truck

for img_name in sorted(os.listdir(input_folder)):
    path = os.path.join(input_folder, img_name)
    img = cv2.imread(path)

    if img is None:
        continue

    # 🔥 run detection with filters
    results = model(
        img,
        conf=0.5,              # confidence threshold
        classes=vehicle_classes,  # ONLY vehicles
        imgsz=640              # consistent input size
    )

    for r in results:
        boxes = r.boxes
        if boxes is None:
            continue

        for box in boxes:
            conf = float(box.conf[0])
            cls = int(box.cls[0])

            x1, y1, x2, y2 = map(int, box.xyxy[0])

            # draw box
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # label
            label = f"{model.names[cls]} {conf:.2f}"
            cv2.putText(img, label, (x1, y1-10),
                        cv2.FONT_HERSHEY_SIMPLEX,
                        0.5, (0,255,0), 2)

    cv2.imwrite(os.path.join(output_folder, img_name), img)

print("🚗 High-accuracy vehicle detection done ✅")
```

```python
import cv2
from ultralytics import YOLO
import math

model = YOLO("yolov8m.pt")

cap = cv2.VideoCapture("/content/input_video.mp4")

fps = cap.get(cv2.CAP_PROP_FPS)

positions = {}   # store history
window_size = 10  # 🔥 number of frames for averaging

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model.track(frame, persist=True, conf=0.5, classes=[2,3,5,7])

    for r in results:
        for box in r.boxes:

            if box.id is None:
                continue

            track_id = int(box.id[0])

            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cx = (x1 + x2) // 2
            cy = (y1 + y2) // 2

            # store positions
            if track_id not in positions:
                positions[track_id] = []

            positions[track_id].append((cx, cy))

            # keep only last N frames
            if len(positions[track_id]) > window_size:
                positions[track_id].pop(0)

            # 🔥 MULTI-FRAME SPEED
            if len(positions[track_id]) == window_size:

                total_dist = 0

                for i in range(1, window_size):
                    p1 = positions[track_id][i-1]
                    p2 = positions[track_id][i]

                    dist = math.sqrt((p2[0]-p1[0])**2 + (p2[1]-p1[1])**2)
                    total_dist += dist

                # time for N frames
                time = window_size / fps

                speed = total_dist / time

                # ⚠️ scale factor (adjust based on camera)
                speed_kmh = speed * 0.05

                # display speed
                cv2.putText(frame, f"{int(speed_kmh)} km/h",
                            (x1, y1-20),
                            cv2.FONT_HERSHEY_SIMPLEX,
                            0.7, (0,255,0), 2)

                # 🚨 overspeed alert
                if speed_kmh > 60:
                    cv2.putText(frame, "OVERSPEED!",
                                (x1, y1-50),
                                cv2.FONT_HERSHEY_SIMPLEX,
                                1, (0,0,255), 3)

            # draw box
            cv2.rectangle(frame, (x1,y1),(x2,y2),(0,255,0),2)

    cv2.imshow("Speed Estimation", frame)

    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import os
import cv2
from ultralytics import YOLO
import math

# Load model
model = YOLO("yolov8m.pt")

input_folder = "/content/detected_frames"
output_folder = "/content/final_frames"
os.makedirs(output_folder, exist_ok=True)

fps = 20
positions = {}
window_size = 10

frame_list = sorted(os.listdir(input_folder))

for frame_idx, img_name in enumerate(frame_list):

    path = os.path.join(input_folder, img_name)
    frame = cv2.imread(path)

    if frame is None:
        continue

    results = model.track(frame, persist=True, conf=0.5, classes=[2,3,5,7])

    for r in results:
        for box in r.boxes:

            if box.id is None:
                continue

            track_id = int(box.id[0])

            x1, y1, x2, y2 = map(int, box.xyxy[0])
            cx = (x1 + x2) // 2
            cy = (y1 + y2) // 2

            # store positions
            if track_id not in positions:
                positions[track_id] = []

            positions[track_id].append((cx, cy))

            if len(positions[track_id]) > window_size:
                positions[track_id].pop(0)

            # default box color = green
            box_color = (0, 255, 0)

            if len(positions[track_id]) == window_size:

                total_dist = 0
                for i in range(1, window_size):
                    p1 = positions[track_id][i-1]
                    p2 = positions[track_id][i]

                    dist = math.sqrt((p2[0]-p1[0])**2 + (p2[1]-p1[1])**2)
                    total_dist += dist

                time = window_size / fps
                speed = total_dist / time

                speed_kmh = speed * 0.5   # 🔥 increased scale

                # show speed
                cv2.putText(frame, f"ID:{track_id} {int(speed_kmh)} km/h",
                            (x1, y1-20),
                            cv2.FONT_HERSHEY_SIMPLEX,
                            0.7, (0,255,0), 2)

                # overspeed alert
                if speed_kmh > 10:
                    print(f"🚨 ALERT ID {track_id} | Speed: {speed_kmh:.2f}")

                    box_color = (0, 0, 255)  # RED

                    cv2.putText(frame, "OVERSPEED!",
                                (50, 80),
                                cv2.FONT_HERSHEY_SIMPLEX,
                                1.5, (0,0,255), 4)

            # draw box ONCE (important)
            cv2.rectangle(frame, (x1,y1),(x2,y2), box_color, 3)

    cv2.imwrite(os.path.join(output_folder, img_name), frame)

print("🚀 All frames processed and saved in /content/final_frames")
```

```python
%%writefile app.py
import streamlit as st
import cv2
import tempfile
from ultralytics import YOLO
import math
import base64
import os

st.set_page_config(layout="wide")
st.title("🚗 Overspeed Detection System")

uploaded_file = st.file_uploader("Upload Video", type=["mp4"])

# ---------------- MODEL ----------------
@st.cache_resource
def load_model():
    return YOLO("yolov8m.pt")

model = load_model()

# ---------------- VIDEO DISPLAY FIX ----------------
def show_video(file_path):
    with open(file_path, "rb") as f:
        video_bytes = f.read()

    b64 = base64.b64encode(video_bytes).decode()

    video_html = f"""
    <video width="100%" controls>
        <source src="data:video/mp4;base64,{b64}" type="video/mp4">
    </video>
    """
    st.markdown(video_html, unsafe_allow_html=True)


# ---------------- MAIN APP ----------------
if uploaded_file is not None:

    tfile = tempfile.NamedTemporaryFile(delete=False, suffix=".mp4")
    tfile.write(uploaded_file.read())
    video_path = tfile.name

    st.video(uploaded_file)

    if st.button("Run Detection"):

        st.info("⏳ Processing started...")

        cap = cv2.VideoCapture(video_path)

        fps = cap.get(cv2.CAP_PROP_FPS)
        if fps == 0 or fps is None:
            fps = 20

        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

        out_path = "output.mp4"
        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        out = cv2.VideoWriter(out_path, fourcc, fps, (width, height))

        positions = {}
        window_size = 10

        frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        progress = st.progress(0)
        status = st.empty()

        frame_idx = 0

        while cap.isOpened():

            ret, frame = cap.read()
            if not ret:
                break

            if frame_idx % 20 == 0:
                status.write(f"Processing frame {frame_idx}/{frame_count}")

            results = model.track(frame, persist=True, conf=0.5, classes=[2,3,5,7])

            if results is None:
                continue

            for r in results:
                if r.boxes is None:
                    continue

                for box in r.boxes:

                    if box.id is None:
                        continue

                    track_id = int(box.id[0])

                    x1, y1, x2, y2 = map(int, box.xyxy[0])
                    cx = (x1 + x2) // 2
                    cy = (y1 + y2) // 2

                    if track_id not in positions:
                        positions[track_id] = []

                    positions[track_id].append((cx, cy))

                    if len(positions[track_id]) > window_size:
                        positions[track_id].pop(0)

                    color = (0, 255, 0)

                    if len(positions[track_id]) == window_size:

                        dist = 0
                        for i in range(1, window_size):
                            p1 = positions[track_id][i-1]
                            p2 = positions[track_id][i]
                            dist += math.dist(p1, p2)

                        time_sec = window_size / fps
                        speed = dist / time_sec
                        speed_kmh = speed * 0.5

                        cv2.putText(frame,
                                    f"ID {track_id}: {int(speed_kmh)} km/h",
                                    (x1, y1 - 10),
                                    cv2.FONT_HERSHEY_SIMPLEX,
                                    0.6, (0,255,0), 2)

                        if speed_kmh > 10:
                            color = (0, 0, 255)
                            cv2.putText(frame,
                                        "OVERSPEED!",
                                        (x1, y1 - 40),
                                        cv2.FONT_HERSHEY_SIMPLEX,
                                        0.8, (0,0,255), 3)

                    cv2.rectangle(frame, (x1,y1),(x2,y2), color, 2)

            out.write(frame)

            frame_idx += 1

            # FIXED PROGRESS BAR (NO CRASH)
            progress.progress(min(frame_idx / frame_count, 1.0))

        cap.release()
        out.release()

        st.success("✅ Processing Completed")

        # ---------------- VIDEO OUTPUT FIX ----------------
        if os.path.exists(out_path) and os.path.getsize(out_path) > 0:
            st.success(f"🎥 Output size: {os.path.getsize(out_path)} bytes")
            show_video(out_path)
        else:
            st.error("❌ Output video not generated properly")
```
